```python
coffee = 10
while True:
  money = int(input("insert money: "))
  if money == 300:
    print("give coffee")
    coffee = coffee - 1
  elif money > 300:
    print("give remainder: %d and give coffee" % (money-300))
    coffee = coffee - 1
  else:
    print("return money and do not give coffee")
    print("leftover coffee amount: %d" % coffee)
  
  if coffee == 0:
    print("no more coffe left. stop selling coffee.")
    break
```
