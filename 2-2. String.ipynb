```python
# String formatting
# 1. Insert number directly
"I eat %d apples." % 3
```

```python
# 2. Insert string directly
"I eat %s apples." % "five"
```

```python
# 3. Insert number by var
num = 3
"I eat %d apples." % num
```

```python
# 4. Insert multiple values
num = 10
day = "three"
"I ate %d apples. so I was sick for %s days" % (num,day)
```

```python
# %s can use any value, because it immediately changes to string
"I have %s apples" % 3
```

```python
"rate is %s" % 3.234
```

```python
# ex) print "Error is 98%"
"Error is %d%%" % 98
```

```python
# common error of above ex
"Error is %d%" % 98
```

```python
# use format code with number
# 1. arrangement, empty space
"%10s" % "hi" # total len = 10 -> hi fills from right side
```

```python
"%-10sjane" % "hi" # total len = 10 -> hi fills from left side -> print jane
```

```python
# 2. express float
"%0.4f" % 3.42134234 # decimal point
```

```python
"%10.4f" % 3.42134234 # decimal point + total len = 10
```
